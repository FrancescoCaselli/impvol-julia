```julia
function in_between(data1, data2, data3)
    # miniscule sign deviations do not count
    return (data1 .- data2) .* (data2 .- data3) .> -2.0
end
function plot_sorted(data)
    sorted = sortrows(data, by = x->x[2])
    r1 = length(sorted[:,1])
    r2 = length(sorted[:,1][in_between(sorted[:,1], sorted[:,2], sorted[:,3])])
    plot(sorted, label=["sigma=0.5", "sigma=1.0", "sigma=2.0"], xlabel="Countries sorted by Cobb-Douglas effect", ylabel="Percentage change in volatility", title="$r2/$r1 corrert orders")
end
```

```julia
# there is something wrong in country 21, needs to be fixed
rows = [1:20; 22:25]
plot_sorted(diversification[rows,:])
```

```julia
plot_sorted(overall)
```
